### c)

Zuerst c. Für unendlich viele Lösungen (oder auch nur zwei) müssen die gleichungen die Selbe Lösungsmenge beschreiben. Daher wird so umgeformt, dass der Faktor vor $x$ gleich ist.


$\begin{align}
x - 5y &= a  | *3\\

\\
\Leftrightarrow 3x - 15y &= 3a \\

\end{align}$

Die verbleibenden Koeffizienten werden verglichen. Es ergibt sich $b = -15$ und $a = \frac{57}{3} = 19$

### b)

Keine Lösung. 

#### Variante eins

$\begin{align}
x - 5y &= a  | *3\\
3x + by &= 57 \\
\\
\Leftrightarrow 3x - 15y &= 3a \\
\Leftrightarrow 3x + by &= 57 | -\\
\\
\Leftrightarrow - 15y - by &= 3a - 57 \\
\\
\Leftrightarrow (- 15 - b)y &= 3a - 57 \\
\end{align}$

Fallunterscheidung $b = -15$

$\begin{align}
(- 15 - (-15))y &= 3a - 57 \\
\Leftrightarrow 0y &= 3a - 57 | + 57 \\
\Leftrightarrow 57 &= 3a | /3 \\
\Leftrightarrow 19 &= a \\
\end{align}$

Für $b = -15$ und $a \neq 19$ widersprechen sich die Gleichungen und es gibt keine Lösung.

Anderer Fall

$\begin{align}
(- 15 - b)y &= 3a - 57 | /(- 15 - b) \\
\Leftrightarrow y &= \frac{3a - 57}{-15- b} \\
\end{align}$

Hier kann $y$ immer berechnet werden und das Gleichungssystem ist lösbar.

#### Variante zwei

Die Lösungsmenge einer linearen Gleichung mit zwei Unbekannten entspricht einer Geraden im zweidimensionalen Raum. Die Lösungsmenge zweier Gleichungen entspricht der Schnittmenge der einzelnen Lösungsmengen. In a) war dies ein einzelner Punkt; in c) waren sie identisch. Damit es keine Lösung gibt, dürfen sich die Geraden nicht schneiden. Sie verlaufen also in die selbe richtung. Aus der Achsenabschnitsform wissen wir, dass die Inhomogenität (die einzelne Zahl rechts) für den Abstand zum Koordinatenursprung verantwortlich ist, also parallel verschiebt. So können wir eine Parallelverschiebung erreichen. Es gilt also $b = -15$ und $a \neq 19$.
