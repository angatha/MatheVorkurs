Es gibt daher folgende Gleichungen mit dem Flächeninhalt $A$, der Seite $c$ und der Höhe über der Seite $c$ als $h_c$.

$\begin{align}
A &= \frac{1}{2}h_cc \\
A + 7 &= \frac{1}{2}(h_c+3)c \\
A - 3 &= \frac{1}{2}(h_c-1)c \\
\end{align}$

Es werden beige gleichungen zu $c$ umgestellt. Es entfällt der nicht lineare Summang $h_cc$ durch subtrahieren auf beiden Seiten.

$\begin{align}
\frac{1}{2}h_cc + 7 &= \frac{1}{2}(h_c+3)c |*2 \\
\frac{1}{2}h_cc - 3 &= \frac{1}{2}(h_c-1)c |*2 \\
\\
h_cc + 14 &= (h_c+3)c \\
h_cc - 6 &= (h_c-1)c \\
\\
h_cc + 14 &= h_cc+3c |-h_cc \\
h_cc - 6 &= h_cc-c |-h_cc \\
\\
14 &= 3c \\
6 &=-c \\
\\
\frac{14}{3} &= c \\
-6 &=c \\
\end{align}$
